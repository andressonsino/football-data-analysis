```python
import pandas as pd
import numpy as np
import os
# import warnings

# Supresión de warnings
# warnings.filterwarnings('ignore')
# pd.options.mode.chained_assignment = None

# Configuración de visualización del DataFrame
pd.set_option('display.max_columns', None)   # muestra todas las columnas
pd.set_option('display.max_rows', 100)
pd.set_option('display.float_format', '{:.2f}'.format)
```

```python
# --- Configuración -------------------------------------------
url = "https://www.transfermarkt.com.ar/club-atletico-huracan/startseite/verein/2063"
dfs = pd.read_html(url)

for i, df in enumerate(dfs):
    print(f"Tabla {i}: shape {df.shape}")
    print(df.head())
    print("-" * 80)
```

```python

# Corregido para leer tablas HTML, no CSV
# read_html devuelve una lista de DataFrames, uno por cada tabla encontrada
dfs_tables = pd.read_html(url)

# Generalmente, la tabla de jugadores es la primera, por lo que usamos el índice [0]
# Ten en cuenta que la tabla devuelta puede necesitar limpieza (renombrado de columnas, etc.)
df_raw = dfs_tables[1]

print(f"Dataset cargado - Filas: {df_raw.shape[0]} | Columnas: {df_raw.shape[1]}")
print(f"✅ Dataset cargado — Filas: {df_raw.shape[0]} | Columnas: {df_raw.shape[1]}")
```

```python
# 1. Guardar el raw tal cual
os.makedirs('data/raw', exist_ok=True)
df_raw.to_csv('data/raw/huracan_plantilla_raw.csv', index=False)
print("✅ Raw guardado")

# 2. Trabajar sobre una copia — NUNCA modificar df_raw directamente
df = df_raw.copy()

# 3. Inspeccionar antes de limpiar
print(df.shape)
print(df.dtypes)
print(df.isnull().sum())
```
